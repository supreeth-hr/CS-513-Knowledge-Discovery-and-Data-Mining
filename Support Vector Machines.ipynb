```python
# Importing Required Libraries and Packages
import pandas as pd
from sklearn import svm
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler
from sklearn.metrics import classification_report, accuracy_score, confusion_matrix
```

```python
df = pd.read_csv("wisc_bc_ContinuousVar.csv") #Loading the dataset
df.head() #Displaying first 5 records
```

```python
#Summary of the dataset
summary = df.describe()
print(summary)
```

```python
#Count the number of missing values in each column of the dataframe
df.isnull().sum() 
```

```python
X = df.drop(columns = ['id','diagnosis'])    #feature(attributes)
y = df['diagnosis']      #target variable
X.head()
```

```python
y.head()
```

```python
# Standardize the feature variables(X)
X_scaled = StandardScaler()
```

```python
X = X_scaled.fit_transform(X)
```

```python
#Converting and ecoding the target variable (y)
y = y.astype("category")
y = y.replace({'B': 0, 'M': 1})
```

```python
y.head()
```

```python
# Split the data into training (70%) and testing (30%) sets
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=123)
```

```python
#Fitting the SVM model
svm_model = svm.SVC(kernel='linear')
svm_model.fit(X_train, y_train)

y_pred = svm_model.predict(X_test)
```

```python
#Computing the accuracy of the model
accuracy = accuracy_score(y_test, y_pred)
print('Accuracy of the SVM model:', accuracy*100, "%")
```

```python
#Computing the confusion matrix and classification report
cm = confusion_matrix(y_test, y_pred)
print('Confusion matrix:')
print(cm)
print('Classification Report')
print(classification_report(y_test, y_pred))
```
