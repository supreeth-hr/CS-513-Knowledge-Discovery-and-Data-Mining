```python
#importing the required libraries
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.naive_bayes import CategoricalNB
from sklearn.metrics import confusion_matrix, classification_report, accuracy_score
```

```python
df=pd.read_csv("breast-cancer-wisconsin.csv") #Loading the dataset
df= pd.DataFrame(df) 
df.head() #Displaying first 5 records
```

```python
#Summary of the dataset
summary = df.describe()
print(summary)
```

```python
df=df[df != '?'].dropna()  # drops NaN values 
print(df[df.isna().any(axis=1)]) # shows NaN values in data frame if it exists
```

```python
# Splitting the dataset
X = df.drop(columns = ['Class'])    #feature(attributes)
y = df['Class']      #target variable
X.head(30)
```

```python
y.head()
```

```python
# Split the data into training (70%) and testing (30%) sets
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=123)
```

```python
# Train a Naïve Bayes classifier (Categorical Naïve Bayes)
nb_model = CategoricalNB()
nb_model.fit(X_train,y_train)
```

```python
# Predict the diagnosis on the test data
predictions = nb_model.predict(X_test)
```

```python
# Compute the confusion matrix
confusion_mat = confusion_matrix(y_test, predictions)

# Compute the classification report
classification_rep = classification_report(y_test, predictions)

# Calculate the accuracy
accuracy = accuracy_score(y_test, predictions)
print("The accuracy of model is ")
print((accuracy*100))

# Calculate the error rate
error_rate = 1 - accuracy

# Print the confusion matrix, classification report, and error rate
print("\n\nConfusion Matrix:")
print(confusion_mat)
print("\nClassification Report:")
print(classification_rep)
print("\nError Rate:", error_rate)
```
