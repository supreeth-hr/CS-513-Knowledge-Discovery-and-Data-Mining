```python
# Importing Required Libraries and Packages
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
from sklearn.tree import DecisionTreeClassifier
from sklearn.model_selection import train_test_split
from sklearn.metrics import accuracy_score
from sklearn.metrics import classification_report
from sklearn.metrics import log_loss, ConfusionMatrixDisplay, average_precision_score, accuracy_score
from sklearn.metrics import roc_curve, precision_recall_curve, auc, f1_score, confusion_matrix
```

```python
df=pd.read_csv("breast-cancer-wisconsin.csv") #Loading the dataset
df= pd.DataFrame(df) 
df.head() #Displaying first 5 records
```

```python
#Describing the data info
summary = df.describe()
print(summary)
```

```python
df=df[df != '?'].dropna()  # drops NaN values 
print(df[df.isna().any(axis=1)]) # shows NaN values in data frame if it exists

# Splitting the dataset
X = df.drop(columns = ['Class','Sample'])    #feature
y = df['Class']      #target variable
X.head()
```

```python
y.head()
```

```python
# Split the dataset into training and testing sets
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=4, shuffle = True)
```

```python
#Defining Decision Tree Model and fitting the model
model = DecisionTreeClassifier()
model.fit(X_train,y_train)
target_pred = model.predict(X_test)
```

```python
#Displaying the Confusion matrix, Accuracy of the model and the Classification Report
print('Confusion Matrix')
print(confusion_matrix(y_test,target_pred))
print('--------------------------------------------------------------')
print('Accuracy score')
print(accuracy_score(y_test,target_pred)*100)
print('--------------------------------------------------------------')
error_rate = 1 - accuracy_score(y_test, target_pred)
print(f"Error Rate: {error_rate:.2f}")
print('--------------------------------------------------------------')
print('Classification Report')
print(classification_report(y_test,target_pred))
```

```python
#Plotting the decision tree
from sklearn.tree import plot_tree

plt.figure(figsize=(50,30), dpi=250)
plot_tree(model,fontsize=20,filled=True,feature_names= X.columns.tolist())
```
