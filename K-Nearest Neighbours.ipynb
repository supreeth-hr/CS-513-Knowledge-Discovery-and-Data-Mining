```python
from sklearn.neighbors import KNeighborsClassifier
from sklearn.model_selection import train_test_split
from sklearn.metrics import accuracy_score
import pandas as pd
```

```python
df=pd.read_csv(r"F:\Chrome Downloads\breast-cancer-wisconsin.csv") #Loading the dataset
df= pd.DataFrame(df) 
df.head() #Displaying first 5 records
```

```python
df=df[df != '?'].dropna()  # drops NaN values 
print(df[df.isna().any(axis=1)]) # shows NaN values in data frame if it exists

# Splitting the dataset
X = df.drop(columns = ['Class'])    #feature
y = df['Class']      #target variable
X.head()
```

```python
y.head()
```

```python
from sklearn.preprocessing import MinMaxScaler
# Initialize MinMaxScaler
scaler = MinMaxScaler()
# Fit and transform the data
X=pd.DataFrame(scaler.fit_transform(X), columns=X.columns)
X.head()
```

```python
# Split the dataset into training and testing sets
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=4, shuffle = True)

k_values = [3, 5, 10] #applying KNN for k=3,5 and 10 and obtaining the accuracy score

for k in k_values:
    knn = KNeighborsClassifier(n_neighbors = k)
    knn.fit(X_train, y_train)
    target_pred = knn.predict(X_test)
    accuracy = accuracy_score(y_test,target_pred)  
    print(f'Accuracy of model with k = {k}: {accuracy} %')
    print('')
```

```python
from sklearn.metrics import classification_report
from sklearn.metrics import confusion_matrix

print('Confusion Matrix')
print(confusion_matrix(y_test,target_pred))
print('--------------------------------------------------------------')
print('Accuracy score')
print(accuracy_score(y_test,target_pred))
print('--------------------------------------------------------------')
error_rate = 1 - accuracy_score(y_test, target_pred)
print(f"Error Rate: {error_rate:.2f}")
print('--------------------------------------------------------------')
print('Classification Report')
print(classification_report(y_test,target_pred))
```
